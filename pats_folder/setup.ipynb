```python
import sys
sys.path.append('../gefera')
import gefera as gf
import numpy as np
import astropy.constants as ac
import matplotlib.pyplot as plt
```

```python
a1 = 124.7 
t1 = 16.56
e1 = 0.00001
p1 = 130.194
w1 = 90.0 * np.pi / 180
b1 = 0.05
i1 = np.arccos(b1 / a1)
a1 = a1 * ac.R_sun.value / ac.au.value

a2 = 61.5 
t2 = 12.41
e2 = 0.00001
p2 = 45.154
w2 = 0.1 * np.pi / 180
om2 = 160 * np.pi / 180
b2 = 0.0
i2 = np.arccos(b2 / a2)
a2 = a2 * ac.R_sun.value / ac.au.value

u1 = 0.6
u2 = 0.2
r1 = 0.1
r2 = 0.07

o1 = gf.PrimaryOrbit(a1, t1, e1, p1, w1, i1)
o2 = gf.ConfocalOrbit(a2, t2, e2, p2, om2, w2, i2)
sys = gf.ConfocalSystem(o1, o2)

```

```python
t=np.linspace(1513.5,1514.1,500)
flux,grad=sys.lightcurve(t,u1,u2,r1,r2,grad=True)
```

```python
plt.plot(t,flux)
```
